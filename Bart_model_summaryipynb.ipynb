```python
from transformers import BartTokenizer, BartForConditionalGeneration
import torch

model_name="facebook/bart-large-cnn"
tokenizer=BartTokenizer.from_pretrained(model_name)
model=BartForConditionalGeneration.from_pretrained(model_name)
```

```python
text="In recent years, climate change has become one of the most pressing global challenges. Rising temperatures, extreme weather events, and melting polar ice caps are just a few of the visible signs of a planet under stress. Governments around the world have been working to reduce greenhouse gas emissions by transitioning to renewable energy sources such as wind and solar. However, progress has been uneven. While some countries have made significant strides, others continue to rely heavily on fossil fuels. Public awareness and activism have also played a key role, with movements like Fridays for Future pushing for urgent action. Scientists warn that unless immediate steps are taken to limit global warming to 1.5 degrees Celsius, the consequences could be catastrophic, including widespread droughts, food shortages, and mass displacement of people."
```

```python
inputs=tokenizer(text,max_length=1024,truncation=True,return_tensors="pt")
```

```python
summary_ids=model.generate(
    inputs["input_ids"],
    max_length=130,
    min_length=30,
    num_beams=4,
    early_stopping=True
)
```

```python
summary=tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("Summary:\n",summary)
```
