```python
sns.barplot(data=t,x="Pclass",y="Survived",palette='coolwarm')
plt.title("Survival rate by passenger class")
plt.xlabel("passenger class")
plt.ylabel("survuval rate")
plt.show()
```

```python
#2)is there any relationship between passenger age and fare?
#create a scatter plot
```

```python
t.dtypes
```

```python
sns.scatterplot(data=t,x=t['Age'],y=t['Fare'],palette='deep')
plt.show("scatter plot for passenger age and fare")
plt.show()
```

```python
#3)what are the pairwise relationships between age ,fare and passenger class>
#pair plot
```

```python
sns.pairplot(data=t[['Age','Fare','Pclass']],palette='coolwarm')
plt.show()
```

```python
numeric_cols=t.select_dtypes(include=['float64','int64'])
cor_mat=numeric_cols.corr()
sns.heatmap(cor_mat,annot=True,cmap="Blues",fmt='.2f')
plt.show()
```
