```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#histogram
sns.histplot(data=tips,x='total_bill',kde=True,bins=20,color="purple")
plt.title("Distribution of total bills")
plt.show()
```

```python
#KDE plot=kernel density estimation plot
```

```python
sns.kdeplot(data=tips,x='total_bill',color="purple",fill=True)
plt.title("Distribution of total bills")
plt.show()
```

```python
#bar plot
```

```python
sns.barplot(data=tips,x='day',y='total_bill',palette='viridis')
plt.show("Bar plot for total bills")
plt.show()
```

```python
#box plot
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue="sex",palette='coolwarm')
plt.show("Bar plot for total bills by day and gender")
plt.show()
```

```python
#scatter plot
#line plot
#pair plot
```

```python
sns.scatterplot(data=tips,x='day',y='tip',hue="day",style='time',palette='deep')
plt.show("scatter plot for tips per day")
plt.show()
```

```python
sns.lineplot(data=tips,x='size',y='tip',marker='o',color='green')
plt.title("line plot of party size vs tip")
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex",palette="husl")
plt.show()
```

```python
#heat map

numeric_cols=tips.select_dtypes(include=['float64','int64'])
cor_mat=numeric_cols.corr()
sns.heatmap(cor_mat,annot=True,cmap="coolwarm",fmt='.2f')
plt.show()
```
