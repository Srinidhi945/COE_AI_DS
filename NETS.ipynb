```python
import os,shutil,pathlib

org_dir=pathlib.Path(r"/content/drive/MyDrive/cat_dog_images/train")
new_base_dir=pathlib.Path(r"/content/drive/MyDrive/image_seg")

def make_subset(subset_name,start_index,end_index):
  for category in ("cat","dog"):
    dir=new_base_dir/subset_name/category
    os.makedirs(dir, exist_ok=True)
    fnames=[f"{category}.{i}.jpg" for i in range(start_index,end_index)]
    for fname in fnames:
      shutil.copyfile(src=org_dir/fname,dst=dir/fname)

make_subset("train",start_index=0,end_index=2000)
make_subset("validation",start_index=2000,end_index=3500)
make_subset("test",start_index=3500,end_index=5500)
```

```python
import cv2
import numpy as np
import oauth2client

labels=['cat','dog']
img_size=224

def get_data(data_dir):
  images=[]
  image_labels=[]
  for label in labels:
    data_path=os.path.join(data_dir,label)
    class_num=labels.index(label)
    for img in os.listdir(data_path):
        try:
            img_arr=cv2.imread(os.path.join(data_path,img))[...,::-1]
            resized_arr=cv2.resize(img_arr,(img_size,img_size))
            images.append(resized_arr)
            image_labels.append(class_num)
        except Exception as e:
            print(f"Error loading image {img}:{e}")
  return np.array(images),np.array(image_labels)
```

```python
train_imgs,train_labels=get_data("/content/drive/MyDrive/image_seg/train")
val_imgs,val_labels=get_data("/content/drive/MyDrive/image_seg/validation")
```
