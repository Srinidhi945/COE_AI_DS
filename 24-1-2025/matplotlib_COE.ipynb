```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.title("line plot")
```

```python
plt.plot(x,y,marker='*',color="red",linestyle="dashed")
```

```python
plt.plot(x,y, marker='*',color='blue',linestyle="dashed",label="line plot",linewidth=3,
        markersize=10,markerfacecolor='red',markeredgecolor='yellow',alpha=0.7)
plt.legend()

#legend to be called in order to show label
```

```python
#create a line plot showing the growth of savings account over 6 month 
a=[1,2,3,4,5,6]
b=[10000,15000,30000,35000,20000,25000]
plt.plot(a,b,marker='*',color="green",linestyle="dashed")
plt.title("TIME(YRS) VS AMOUNT")
plt.xlabel("time in years")
plt.ylabel("savings amount in Rs")
```

```python
a=[1,2,3,4,5,6]
b=[10000,15000,30000,35000,20000,25000]
plt.plot(a,b,marker='*',color="green",linestyle="dashed")
plt.title("TIME(YRS) VS AMOUNT")
plt.xlabel("time in years")
plt.ylabel("savings amount in Rs")
plt.grid(True)

#line chart
```

```python
categories =['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color="crimson")
plt.title("basic bar chart")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()

#bar chart
```

```python
plt.bar(categories,values,color={"crimson","blue","yellow","orange"})
plt.title("basic bar chart")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35,45,46,43,55,53,57,58,56,56,67,65,67,68]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color="skyblue",edgecolor="black")
plt.title("histogram")
plt.xlabel("age ranges")
plt.ylabel("frequency")
plt.show()

#histogram
```

```python
categories =['A','B','C','D','E']
values=[10,15,7,12,9]
plt.pie(values,labels=categories)
plt.title("basic pie chart")
plt.show()
```

```python
prg_lang=['python','c','java']
stds=[45,25,30]
plt.pie(stds,labels=prg_lang,colors=["crimson","purple","yellow"],autopct='%1.1f%%')
plt.title("programming languages vs students interest")
plt.show()
```

```python
d=[10,20,30,40,50]
e=[20,40,60,80,100]
plt.scatter(d,e,color=["crimson","purple","yellow","cyan","teal"],s=100,edgecolor='black')
plt.title("scatter plot")
plt.xlabel('category')
plt.ylabel('values')
plt.show()
```
