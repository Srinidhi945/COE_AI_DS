```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

```

```python


data = pd.read_csv(r'C:\Users\CVR\Downloads\marketing_campaign.csv', delimiter='\t')


print(data.head())

```

```python
print(data.info())  # Check data types of all columns
```

```python
data.isnull().sum()
```

```python
# Fill missing values with the column's mean
data['Income'] = data['Income'].fillna(data['Income'].mean())
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical columns
label_enc = LabelEncoder()
data['Education'] = label_enc.fit_transform(data['Education'])
data['Marital_Status'] = label_enc.fit_transform(data['Marital_Status'])
```

```python
import datetime as dt

# Convert 'Dt_Customer' to datetime, specifying day-first format
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'], dayfirst=True)

# Calculate days since enrollment
data['Days_Enrolled'] = (dt.datetime.now() - data['Dt_Customer']).dt.days

# Drop the original 'Dt_Customer' column
data.drop('Dt_Customer', axis=1, inplace=True)

```

```python
print(data[['Days_Enrolled']].head())
```

```python
features = [
    'Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
    'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
    'NumStorePurchases', 'NumWebVisitsMonth', 'Kidhome', 'Teenhome',
    'Education', 'Marital_Status', 'Days_Enrolled'
]
X = data[features]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Apply Agglomerative Clustering
agglomerative = AgglomerativeClustering(n_clusters=3)  # Adjust 'n_clusters' based on your needs
labels = agglomerative.fit_predict(X_scaled)

# Add cluster labels to the DataFrame
data['Cluster'] = labels

# Visualize clusters (if data is 2D)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis')
plt.title('Agglomerative Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

# Compute linkage matrix
Z = linkage(X_scaled, method='ward')  # Choose 'ward', 'complete', or 'average'

# Plot dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Divisive Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

# Compute linkage matrix
Z = linkage(X_scaled, method='complete')  # Choose 'ward', 'complete', or 'average'

# Plot dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Divisive Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
