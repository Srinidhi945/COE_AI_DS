```python
from transformers import AutoTokenizer,AutoModelForTokenClassification
from transformers import TokenClassificationPipeline
import torch

model_name="dslim/bert-base-Ner"
tokenizer=AutoTokenizer.from_pretrained(model_name)
model=AutoModelForTokenClassification.from_pretrained(model_name)

ner_pipeline=TokenClassificationPipeline(model=model,tokenizer=tokenizer,aggregation_strategy="simple")
```

```python
text="Apple Inc. is planning to open a new office in Berlin by March 2026, and Tim Cook is expected to attend the inauguration."
entities=ner_pipeline(text)

for ent in entities:
  print(f"{ent['word']}  -> {ent['entity_group']} (score:{ent['score']:.3f})")
```
