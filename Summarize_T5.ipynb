LOAD MODEL

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch

model_name="t5-base"
tokenizer=T5Tokenizer.from_pretrained(model_name)
model=T5ForConditionalGeneration.from_pretrained(model_name)
```

INPUT TEXT FOR SUMMARIZATION

```python
text="""
The monarch butterfly, Danaus plexippus, embarks on one of the most incredible migrations in the insect world, traveling
 thousands of miles from Canada and the United States to overwintering sites in Mexico and California. This arduous journey
  is not undertaken by a single generation; rather, it spans multiple generations, with the final generation, known as the "super generation,
  " making the entire southward trek. These butterflies navigate using a sophisticated internal compass, sensitive to the Earth's magnetic field
  and the position of the sun. Their survival is intrinsically linked to the availability of milkweed, the sole host plant for their larvae, and the
   preservation of their overwintering habitats, which are increasingly threatened by deforestation and climate change. Conservation efforts are crucial
    to ensuring the continued existence of this iconic species and its remarkable migratory phenomenon.
    """
```

PREPEND THE TASK PREFIX "SUMMARIZE:"

```python
input_text="summarize:"+text
```

TOKENIZE INPUT

```python
inputs=tokenizer.encode(input_text,return_tensors="pt",max_length=512,truncation=True)
```

GENERATE SUMMARY

```python
summary_ids=model.generate(
    inputs,
    max_length=100,
    min_length=30,
    length_penalty=2.0,
    num_beams=4,
    early_stopping=True
)
```

DECODE AND PRINT

```python
summary=tokenizer.decode(summary_ids[0],skip_special_tokens=True)
print("summary:\n",summary)
```
